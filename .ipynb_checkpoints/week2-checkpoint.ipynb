### Task 2.2

```python
# Корни полинома: -1,8; -0,8; 2

roots = [-1.8, -0.8, 2]
coefs = np.poly(roots)
print("Coefficients of the polynomial: ", coefs)

px = np.poly1d(coefs)
print (px) 
```

```python
# px = x^3 + 0.6*x^2 - 3.76*x - 2.88

x_list = roots
y1_list = list()
y2_list = list()

for x in range(len(x_list)):
    y1 = (math.sin(5*roots[x]))/(5*roots[x])
    y2= -0.2*(roots[x]**3 + 0.6*roots[x]**2 - 3.76*roots[x] - 2.88)
    y1_list.append(y1)
    y2_list.append(y2)

plt.plot(x_list, y1_list, linestyle='dashdot', label='y1 graph')
plt.plot(x_list, y2_list, linestyle='dotted', label='y2 graph' )
plt.legend(loc='lower right', fontsize=15)
```

### Task 2.3

```python
N = 300

x = np.random.rand(N)

y1 = np.random.normal(4, 2, size=N)
y2 = np.random.uniform(low=0, high=7.5, size=N)
y3 = np.random.gamma(shape=2, scale=1.7, size=N)

ax = plt.axes()
ax.set_facecolor("black")

plt.scatter(x, y1, color=(1, 1, 1), s=1)
plt.scatter(x, y2, color='b', s=2)
plt.scatter(x, y3, color='#FF0000', s=3)
```

### Task 2.4

```python
# Амплитуды колебаний
A = 10
B = 15
# Частоты
a = 2
b = 4
# Сдвиг фаз
g = math.pi

x_list = list();
y_list = list();

# Lissajous curve
for i in range(0, 180):
    t = i * 0.02
    xt = A*math.sin(a*t + g)
    yt = B*math.sin(b*t)
    x_list.append(xt)
    y_list.append(yt)

#plt.plot(1, 1, linewidth=3)
plt.title("Lissajous Curve", fontsize=18)
plt.plot(y_list, x_list, color=(0.5, 0.3, 0.7))

plt.grid(color = 'black')

plt.show()
```
